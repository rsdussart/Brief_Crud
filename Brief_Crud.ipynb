```python
populations_data = pd.read_csv('country.csv',index_col=False, delimiter = ',',encoding='UTF-8')
populations_data.head()
rows = populations_data.iloc[:,[0,5,6]]
rows.head()

rows = pd.DataFrame(rows,columns=['ID','LOCATION','TIME','Value'])
print(rows) 
for index, rows in populations_data.iterrows():
    query = '''INSERT INTO populations (ID,LOCATION,TIME,Value) VALUES (%s, %s, %s, %s) ON DUPLICATE KEY UPDATE LOCATION = VALUES(LOCATION), TIME = VALUES(TIME), Value = VALUES(Value)'''
    time_string = str(rows['TIME'])
    date_object = datetime.strptime(time_string, '%Y')
    mysql_date = date_object.strftime('%Y-%m-%d')
    values = (index, rows['LOCATION'], mysql_date,rows['Value'])
    mycursor.execute(query, values)

# commit the changes to the database
mydb.commit()

# close the cursor and connection
mycursor.close()
mydb.close()
```

```python
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="example",
    port="3307",
    database="CountriesDatas_II"
)
mycursor = mydb.cursor()
# check the connection
if mydb.is_connected():
    print("connected")
else:
    print("not connected")
    
# SELECT query with JOIN
query = '''SELECT populations.ID, countries.name, populations.TIME, populations.Value FROM populations JOIN countries ON populations.LOCATION = countries.code'''
mycursor.execute(query)

# fetch the results
results = mycursor.fetchall()

# print the results
for result in results:
    print(result)

# close the cursor and connection
mycursor.close()
mydb.close()

#Si cela ne résout pas le problème, vous pouvez également vérifier les informations de votre fichier de configuration de mysql (my.cnf) pour vous assurer qu'il n'y a pas de limite de connection.
```
